```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
```

```python
def get_nba_player_stats():
    # past week, modify if needed
    end_date = datetime.today().date()
    start_date = end_date - timedelta(days=7)

    start_date_str = start_date.strftime('%Y%m%d')
    end_date_str = end_date.strftime('%Y%m%d')
    
    url = f'https://www.espn.com/nba/stats/player/_/season/2023/seasontype/2/count/{start_date_str}-{end_date_str}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='Table Table--align-right')
    df = pd.read_html(str(table))[0]

    return df
```

```python
def get_nba_names():
    # past week, modify if needed
    end_date = datetime.now()
    start_date = end_date - timedelta(days=7)
    
    start_date_str = start_date.strftime('%Y%m%d')
    end_date_str = end_date.strftime('%Y%m%d')
    
    url_template = "https://www.espn.com/nba/stats/player/_/table/offensive/sort/points/dir/desc/qualified/false/count/{count}/from/{start_date}/to/{end_date}"
    
    # player variables
    players = []
    count = 1
    has_more_data = True
    
    # loop through pages
    while has_more_data:
        url = url_template.format(count=count, start_date=start_date_str, end_date=end_date_str)
        

        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
    
        table = soup.find('table', class_='Table')
        
        if not table:
            break
        
        # player names and positions
        for row in table.tbody.find_all('tr'):
            cells = row.find_all('td')
            name = cells[1].text.strip()
            position = cells[0].text.strip()
            players.append({'name': name, 'position': position})
        
        # next page of data check
        has_more_data = soup.find('a', class_='AnchorLink', string='Next')
        count += 50
    
    return pd.DataFrame(players)

df_names = fetch_nba_stats()
print(df_names)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://www.espn.com/nba/salaries"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table", class_="tablehead")


rows = table.find_all("tr")
header = [th.get_text() for th in rows[0].find_all("td")]

# salary data
data = []
for row in rows[1:]:
    data_row = [td.get_text() for td in row.find_all("td")]
    data.append(data_row)

df_salary = pd.DataFrame(data, columns=header)

print(df_salary)
```
